# Part 1

Writing a function for $f(x) = x^3 - x^2 - 1$

```python
def f(x):
    """
    This function performs a set of operations on an input x, namely:
    
    x^3 - x^2 - 1
    
    Parameters
    ----------
    x : float
        Input
        
    Returns
    -------
    float
        Performed operation on x
    """
    
    return (x ** 3) - (x ** 2) - 1

def df(x):
    """
    This function is the derivative of f(x) with respect to x. Computed to be:
    
    3x^2 - 2x
    
    Parameters
    ----------
    x : float
        Input
        
    Returns
    -------
    float
        derivative of f wrt x at given input value.
    """
    return (3 * x**2) - (2 * x)
```

# Part 2

Newton iteration

```python
def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    """
    This function performs the newton iteration on function f. 
    It finds the root y such that f(y)=0 through recursion with:
    
    y_(n+1) = y_n - f(y_n)/df(y_n)
    
    Parameters
    ----------     
    f : function
        victim of newton iteration
    df : function
        derivative of f with respect to input value
    x0 : float
        initial input value
    epsilon : float
        if the magnitude of f(x) is less than this, it is considered to be 0
    max_iter : int
        the maximum number of iterations allowed
    
    Returns
    -------
    float
        A float value of x_n if the operation succeeds
    err_val : None
        if n exceeds the maximum iteration value
    """
    
    x = x0
    n = 0
    while n < max_iter:
        if abs(f(x)) < epsilon:
            print(f'Found root in {n} iterations.')
            return x
        else:
            x = x - (f(x)/df(x))
            n += 1
    
    print(f'Iteration failed')
    return None
```
