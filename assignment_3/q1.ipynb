```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
from q1_iterator import complex_iterate, check_convergence

colours = cm.inferno(np.linspace(0, 0.9, 10))
```

```python
x = np.linspace(-2, 2, 1000) # Real part of c
y = np.linspace(-2, 2, 1000) # Imaginary part of c
X,Y = np.meshgrid(x,y,indexing='ij')

z = [[0 for _ in y] for _ in x] # Initialising an empty array

for a in range(len(x)):
    for b in range(len(y)):
        z[a][b] = check_convergence(complex_iterate(x[a], y[b]))

Z = np.array(z)

plt.figure(dpi=100)
plt.contourf(X, Y, Z, levels=50, cmap=cm.inferno)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.colorbar(label="-1 = convergence, 0 = divergence")
plt.title(r'Convergence of $c=x+iy$')
plt.show()
```

```python
k = [[0 for _ in y] for _ in x] # Initialising an empty array

for a in range(len(x)):
    for b in range(len(y)):
        k[a][b] = complex_iterate(x[a], y[b])

K = np.array(k)

plt.figure(dpi=100)
plt.contourf(X, Y, K, levels=31, cmap=cm.inferno)
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="Iteration # on divergence")
plt.title(r'Convergence of $c=x+iy$')
plt.show()
```
