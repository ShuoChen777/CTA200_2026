# Breakdown

20 rounds per game, 2 bots against one another. Bot behaviours:

- Always cooperate
- Always defect
- Random
- Resentful
- Tit for Tat

```python
import random
```

```python
# I'll make classes for each type of bot
class Bot():
    """
    Basic bot class for shared functions for all other bots
    """
    def __init__(self):
        self.points = 0
        self.moveset = []
        self.name = 'Bot'
    
    def choice(self, enemy_turn=[]):
        pass
    
    def add_points(self, points):
        self.points += points
        
    def add_move(self, move):
        self.moveset.append(move)
    

class AlwaysCooperate(Bot):
    """
    Bot will always cooperate on its turn.
    """
    
    def __init__(self):
        super().__init__()
        self.name = 'Cooperative'
    
    def choice(self, enemy_turn=[]):
        return 'c' # stands for cooperate


class AlwaysDefect(Bot):
    """
    Bot will always defect on its turn.
    """
    
    def __init__(self):
        super().__init__()
        self.name = 'Defective'
    
    def choice(self, enemy_turn=[]):
        return 'd' # stands for defect
    

class Random(Bot):
    """
    Bot will make random decisions
    """
    
    def __init__(self):
        super().__init__()
        self.name = 'Random'
    
    def choice(self, enemy_turn=[]):
        if random.randint(1, 2) == 2:
            return 'c'
        else:
            return 'd'
        
        
class Resentful(Bot):
    """
    Bot will defect if its enemy ever defects
    """
    
    def __init__(self):
        super().__init__()
        self.name = 'Resentful'
    
    def choice(self, enemy_turn=[]):
        if 'd' in enemy_turn:
            return 'd'
        else:
            return 'c'

    
class Tittat(Bot):
    """
    Bot will play last enemy move
    """
    
    def __init__(self):
        super().__init__()
        self.name = 'Tit-for-tat'
    
    def choice(self, enemy_turn=[]):
        if len(enemy_turn) == 0:
            return 'c'
        return enemy_turn[-1]
```

```python
# I'll construct a function for the game

def tournament(class1, class2, rounds=20):
    bot1 = class1()
    bot2 = class2()
    print(f'Bot 1 is {bot1.name}')
    print(f'Bot 2 is {bot2.name}')
    for round in range(rounds):
        # Both bots make a move
        choice = {'bot1': bot1.choice(bot2.moveset),
                  'bot2': bot2.choice(bot1.moveset)}
        
        # Here we choose which bot gets how many points
        if choice['bot1'] == choice['bot2']:
            if choice['bot1'] == 'd':
                bot1.add_points(1)
                bot2.add_points(1)
            else:
                bot1.add_points(3)
                bot2.add_points(3)
        else:
            if choice['bot1'] == 'd':
                bot1.add_points(5)
            else:
                bot2.add_points(5)
        
        # Append the move to each bot's moveset
        bot1.add_move(choice['bot1'])
        bot2.add_move(choice['bot2'])
    
        # Every 5 rounds, display score
        if round % 5 == 0:
            print(f'Round {round + 1} score:\nBot 1 {bot1.points} - {bot2.points} Bot 2')
    
    print(f'Final score:\nBot 1 {bot1.points} - {bot2.points} Bot 2')
```

```python
tournament(Resentful, Random)
```

```python
tournament(Resentful, Tittat)
```
